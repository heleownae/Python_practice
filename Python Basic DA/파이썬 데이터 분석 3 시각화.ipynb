3. 그래프 그리기 - 차트

(1) Line

```python
data = sns.load_dataset('flights')
data
```

```python
# 연도별 총 승객수의 변화 추이

data_grouped = data[['year', 'passengers']].groupby('year').sum().reset_index()
data_grouped

plt.plot(data_grouped['year'], data_grouped['passengers'])
plt.xlabel('year')
plt.ylabel('passengers')
plt.show()
```

(2) Bar vs Histogram

```python
df = pd.DataFrame({
    '도시': ['서울', '부산', '대구', '인천'],
    '인구': [990, 250, 250, 290]
})
df
```

```python
# Bar (막대그래프)
plt.bar(df['도시'], df['인구'])
plt.xlabel('도시')
plt.ylabel('인구')
plt.title('도시별 인구 수')
plt.show()
```

```python
# Histogram

# 랜덤하게 1000개의 데이터를 생성
data = np.random.randn(1000)
data
```

```python
# Histogram

# bins: 구간별로 여러 데이터가 있을 때, 몇 개의 구간을 둬서 데이터를 표현할 건지 '구간 값'
plt.hist(data, bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram_Example')
plt.show()
```

(4) Box plot

```python
iris = sns.load_dataset("iris")
iris
```

```python
species = iris['species'].unique()
```

```python
sepal_length_list = [iris[iris['species'] == s]['sepal_length'].tolist() for s in iris['species'].unique()]
sepal_length_list
```

```python
len(sepal_length_list[2])
```

```python
plt.boxplot(sepal_length_list, labels=species)
plt.xlabel('Species')
plt.ylabel('Sepal Length')
plt.title("Box Plot")
plt.show()

# 가운데 노란 선 = 중앙값
# 중앙값 기준 아래 = 25% / 위 = 75%
# 상하에 뻗은 가로선 = 최대값, 최솟값
# 오른쪽 하단 점 = 이상치(Outlier)
```

(5) Boxplot 2 (Seaborn 활용)

```python
sns.boxplot(x='species', y='sepal_length', data=iris)
plt.show()
```
