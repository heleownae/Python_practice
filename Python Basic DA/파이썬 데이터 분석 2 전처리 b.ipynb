1. 데이터 병합

```python
# 데이터프레임 만들기

df1 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3']
})

df2 = pd.DataFrame({
    'A': ['A4', 'A5', 'A6', 'A7'],
    'B': ['B4', 'B5', 'B6', 'B7'],
    'C': ['C4', 'C5', 'C6', 'C7'],
    'D': ['D4', 'D5', 'D6', 'D7']
})

df3 = pd.DataFrame({
    'A': ['A8', 'A9', 'A10', 'A11'],
    'B': ['B8', 'B9', 'B10', 'B11'],
    'C': ['C8', 'C9', 'C10', 'C11'],
    'D': ['D8', 'D9', 'D10', 'D11']
})

df4 = pd.DataFrame({
    'E': ['E0', 'E1', 'E2', 'E3'],
    'F': ['F0', 'F1', 'F2', 'F3'],
    'G': ['G0', 'G1', 'G2', 'G3'],
    'H': ['H0', 'H1', 'H2', 'H3']
})

df5 = pd.DataFrame({
    'A': ['A0', 'A1', 'A2', 'A3'],
    'B': ['B0', 'B1', 'B2', 'B3'],
    'C': ['C0', 'C1', 'C2', 'C3'],
    'D': ['D0', 'D1', 'D2', 'D3'],
    'E': ['E0', 'E1', 'E2', 'E3']
})
```

(1) concat (단순 병합)

```python
# 각 df를 세로(위아래)로 붙이고, 인덱스 리셋

# axis=0 이 기본 세로로 붙이기임, 굳이 하지않더라도 위아래로 붙여짐

pd.concat([df1, df2, df3]).reset_index(drop=True)
pd.concat([df1, df2, df3], axis=0).reset_index(drop=True) 
```

```python
# 각 df를 가로(양 옆)로 붙이기

pd.concat([df1, df4], axis=1)
```

```python
# 만약 각 df 컬럼 수가 딱 맞지 않다면? Null로 채워짐

pd = pd.concat([df5, df2, df3]).reset_index(drop=True)
pd
```

(2) merge (특정 컬럼을 고려한 병합 same as SQL JOIN)

```python
import pandas as pd

left_df = pd.DataFrame({'key': ['A', 'B', 'C', 'D'], 'value': [1, 2, 3, 4]})
right_df = pd.DataFrame({'key': ['B', 'D', 'E', 'F'], 'value': [5, 6, 7, 8]})
```

```python
# key 컬럼 기준으로 outer 조인 ㅎㅅㅎ
pd.merge(left_df, right_df, on='key', how='outer')
```

```python
# key 컬럼 기준으로 left 조인 ㅎㅅㅎ
pd.merge(left_df, right_df, on='key', how='left')
```

```python
# key 컬럼 기준으로 right 조인 ㅎㅅㅎ
pd.merge(left_df, right_df, on='key', how='right')
```

2. 데이터 집계

(1) group by

```python
data = {
    'Category': ['A', 'B', 'A', 'B', 'A', 'B'],
    'Value': [1, 2, 3, 4, 5, 6]
}

df = pd.DataFrame(data)
df
```

```python
# 평균
df.groupby('Category').mean()
```

```python
# 합
df.groupby('Category').sum()
```

```python
# 개수
df.groupby('Category').count()
```

```python
# 최대값 
df.groupby('Category').max()
```

```python
# 최솟값
df.groupby('Category').min()
```

```python
# 첫 번째 값
df.groupby('Category').first()
```

```python
# agg - 다중집계작업을 간결하게 나타냄
df.groupby('Category').agg(list)
```

(2) group by 실습

```python
data = sns.load_dataset('tips')
data

# 데이터 저장하기
data.to_csv("tips_data.csv", index=False)
df = pd.read_csv("tips_data.csv")

df
```

```python
df[['day', 'total_bill', 'tip', 'size']].groupby('day').mean()
```

```python
df[['sex', 'day', 'total_bill', 'tip', 'size']].groupby(['sex', 'day']).mean()
```

```python
# 각각의 컬럼에 다른 집계함수를 쓸 때
df[['sex', 'day', 'total_bill', 'tip', 'size']].groupby(['sex', 'day']).agg({'total_bill': 'max', 'tip': 'mean', 'size': 'sum'})
```

(3) 피봇 테이블

```python
# 샘플 데이터프레임1 생성
data = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-01'],
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'Value': [10, 20, 30, 40, 50]
}
df = pd.DataFrame(data)
```

```python
# 피봇 테이블 만들기

pivot = df.pivot_table(index='Date', columns='Category', values='Value', aggfunc='sum')
pivot
```

```python
# 샘플 데이터프레임2 생성
data2 = {
    'Date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02', '2023-01-01'],
    'Category': ['A', 'B', 'A', 'B', 'A'],
    'SubCategory': ['X', 'X', 'Y', 'Y', 'Z'],
    'Value': [10, 20, 30, 40, 50]
}
df2 = pd.DataFrame(data2)
df2
```

```python
# 피봇 테이블 만들기

pivot2 = df2.pivot_table(index='Date', columns=['Category', 'SubCategory'], values='Value', aggfunc='sum')
pivot2
```

3. 데이터 정렬

```python
# 샘플 데이터프레임 생성
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 22, 30, 30, 27],
    'Score': [85, 90, 75, 80, 95]
}
df = pd.DataFrame(data)
df
```

```python
# Age로 오름차순 정렬하기
df.sort_values(by='Age')
```

```python
# Age로 내림차순 정렬하기
df.sort_values(by='Age', ascending=False)
```

```python
# Age 오름차순, Score로 내림차순 정렬하기
# 먼저 입력한 컬럼부터 정렬됨

df.sort_values(by=['Age', 'Score'], ascending=[True, False])
```

```python
# 인덱스로 정렬

df.sort_index(ascending=False)
```
