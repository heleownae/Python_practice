4. 데이터 선택하기

```python
# .iloc (Integer Location) - 숫자 타입의 원하는 행, 열 선택하기

#데이터 생성
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
})

df
```

```python
# .iloc (Integer Location) - 숫자 타입의 원하는 행, 열 선택하기

df.iloc[0:2]   # 0번째 로우부터 총 2개 = 인덱스0 로우 ~ 인덱스1 로우까지
```

```python
# iloc - 슬라이싱
df.iloc[0:5:2] # 0번째 로우부터 총 5개, 2개씩 건너뜀
               # 인덱스0 로우, 인덱스2 로우, 인덱스4 로우

df.iloc[0::2]  # 0번째 로우부터 끝까지, 2개씩 건너뜀
```

```python
# iloc - 행, 열 콕 집어서 추출!!

df.iloc[0, 0:2]  # 인덱스0 행의, 인덱스0~1 열
```

```python
# loc (Location) - 숫자가 아닌 타입의 원하는 행, 열 선택하기

#데이터에 인덱스 생성
df = pd.DataFrame({
    'A': [1, 2, 3, 4, 5],
    'B': [10, 20, 30, 40, 50],
    'C': [100, 200, 300, 400, 500]
}, index=['a', 'b', 'c', 'd', 'e'])

df
```

```python
# 전체 데이터에서 첫 번째 컬럼만 보기 - iloc
df.iloc[:, 0]
```

```python
# 전체 데이터에서 첫 번째 컬럼만 보기 - loc
df.loc[:, 'A']   # 컬럼명을 직접 입력
```

```python
# loc 슬라이싱 활용 사례

df.loc['b':'d']
df.loc['b':'d', 'A':'C']
```

```python
# 슬라이싱 & 선택

df.loc[:, 'A']
df['A']
```

```python
# 슬라이싱 & 선택

# df['A', 'B']  # 에러
df[['A', 'B']]  # 에러 해결 - 리스트로 묶기
```

5. 불리언 인덱싱: 내가 입력한 조건이 '참(True)'인 경우만 불러오기

```python
df = pd.read_csv("tips_data.csv")
df
```

```python
# 남성인 경우

df[df['sex']=='Male']
df[df['sex']!='Female']
```

```python
# 남성 흡연자인 경우

df[(df['sex']=='Male') & (df['smoker']=='Yes')]
```

```python
# 남성이거나 흡연자인 경우

df[(df['sex']=='Male') | (df['smoker']=='Yes')]
```

```python
# loc 활용하기 - size가 3 초과인 경우

df.loc[df['size']>3, 'tip':'smoker']
```

```python
# isin 활용하기 - 1, 2 사이즈인 경우

df[df['size'].isin([1, 2])]
```

```python
# isin 활용하기 - sun, thurs인 경우

df[df['day'].isin(['Sun', 'Thur'])]
```

```python
# 실습 1 

condition = df['tip'] < 2
df[condition]
```

```python
# 실습 2 - 조건이 여러개이며, 각각 변수 지정

cond1 = df['size'] >= 3
cond2 = df['tip'] < 2

df[cond1 & cond2]
```

```python
# 실습 3 - 조건이 여러개이며, 하나의 변수 지정

cond3 = (df['size'] >= 3) & (df['tip'] < 2)

df[cond3]
```
